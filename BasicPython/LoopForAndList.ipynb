### for (continue - break)

```python
segredo = 'python3'
chances = 3

while True:
    letra = input('Digite uma letra: ')
    
    if chances <= 0:
        print('vc perdeu')
        break
        
    else:
        
        if len(letra) > 1:
            print('Isso nao vale, digite apenas uma letra!')
            continue

        if letra in segredo:
            print('A letra: {} existe na palavra secreta'.format(letra))
            break
            
        elif letra not in segredo:
            chances -= 1
            print('A letra: {} existe nao existe palavra secreta'.format(letra))
            
```

```python
var = ['A', 'B', 'C']

for i in var:
    print(i)
    print(i)
```

```python
## toda vez que acha o continue ele vai pra iteracao do  loop
for i in var:
    print(i)
    continue
    print(i)
```

```python
## assim que acha o break ele vai sair do loop
for i in var:
    print(i)
    break
```
