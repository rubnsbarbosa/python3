```python
'''
+  - adição
-  - subtração
*  - multiplicação
/  - divisão
// - divisão inteira - o resultado é arredondado
** - exponenciação
%  - retorna o módulo - resto da divisão de um número por outro
() - altera a precedencia entre números
'''

print(10 + 10)
print(10 - 2)
print(10 * 10)
print(10 / 2)
print(10.5 / 2)
print(10.5 // 2)
print(2 ** 8)
print(4 % 2)
print(4 % 3)
print((4 + 1) * 2)
print(4 * 1 + 2)
```
