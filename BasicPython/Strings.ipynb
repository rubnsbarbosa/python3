### Interando strings com while

```python
minha_string = 'o rato roeu a roupa do rei de roma'
## string é imutável em python
```

```python
print(minha_string[2:6])
```

```python
c = 0
tamanho_string = len(minha_string)

while c < tamanho_string:
    print(minha_string[c])
    c += 1
```

```python
## pra eu mudar os r para R 
nova_string = ''
ch = 0
while ch < tamanho_string:
    if minha_string[ch] == 'r':
        nova_string += minha_string[ch].upper()
    else:
        nova_string += minha_string[ch]
    
    ch += 1
    
print(nova_string)
```
