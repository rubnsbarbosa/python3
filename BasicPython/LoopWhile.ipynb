```python
'''
While em python
Utilizado para realizar ações enquanto uma condição for verdadeira.
Repetições
'''
a = 1

while a <= 5:
    print(a)
    a = a + 1
```

```python
linha = 0
coluna = 0

while linha < 10:
    coluna = 0
    while coluna < 5:
        print()
        coluna += 1
    linha += 1
```

```python
contador = 1
acumulador = 1

while contador <= 10:
    print(contador)
    
    if contador > 5:
        break
    
    acumulador = acumulador + contador
    contador += 1
else:
    print('cheguei no else')
print('Acumulador = ' + str(acumulador))
```
